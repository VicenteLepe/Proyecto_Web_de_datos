```python
import pandas as pd

# 1️⃣ Cargar el CSV original
df = pd.read_csv("games-data.csv")

# 2️⃣ Eliminar filas donde la columna "user score" sea "tbd"
df = df[df["user score"] != "tbd"]

# 3️⃣ Convertir la columna 'r-date' a datetime (formato ISO)
df["r-date"] = pd.to_datetime(df["r-date"], format="%B %d, %Y", errors="coerce")

# 4️⃣ Formatear la fecha como 'YYYY-MM-DD'
df["r-date"] = df["r-date"].dt.strftime("%Y-%m-%d")

# Suponiendo que la columna se llama 'genre'
df = df.assign(genre=df['genre'].str.split(',')) \
       .explode('genre')

# Quitar espacios extra
df['genre'] = df['genre'].str.strip()

# 5️⃣ Guardar el CSV limpio
df.to_csv("games-data-clean.csv", index=False)



print("✅ Archivo 'games-data-clean.csv' generado correctamente.")
```
